Ini Adalah data penjualan bulanan sebuah toko selama setahun

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data penjualan (bulan dan jumlah penjualan dalam juta rupiah)
data = {
    'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'],
    'Penjualan': [120, 135, 150, 145, 160, 155, 180, 200, 190, 210, 220, 250]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Hitung total penjualan setahun
total_penjualan = df['Penjualan'].sum()
print(f'Total Penjualan Setahun: {total_penjualan} juta rupiah')

# Visualisasi tren penjualan
plt.figure(figsize=(10,6))
plt.plot(df['Bulan'], df['Penjualan'], marker='o', linestyle='-', color='b')
plt.title('Tren Penjualan Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan (juta rupiah)')
plt.grid(True)
plt.show()
```

Interpretasi Hasil :
- Total Penjualan:
Program ini menghitung total penjualan selama setahun. Dari Contoh data yang Ada:
120 + 135 + 150 + 145 + 160 + 155 + 180 + 200 + 190 + 210 + 220 + 250 = 2115 juta rupiah.
Artinya, total penjualan setahun mencapai Rp2,115 miliar.

Visualisasi Tren Penjualan:
 - Grafik garis yang dihasilkan akan memperlihatkan tren penjualan bulanan.
 - Bulan awal (Januari-Maret) menunjukkan penjualan stabil meningkat.
 - Terjadi sedikit fluktuasi pada bulan April-Mei.
 - Lonjakan penjualan terjadi mulai Juli, dengan puncak tertinggi di bulan Desember (250 juta).
 - Hal ini mungkin karena momen belanja besar seperti liburan akhir tahun.
